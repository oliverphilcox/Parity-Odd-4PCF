```python
%pylab inline
from nbodykit import cosmology, transform
import os
```

```python
cosmo = cosmology.Cosmology(Omega_cdm=0.260448,Omega_b=0.022/0.676**2.,h=0.676,sigma8=0.8,n_s=0.96,m_ncdm=0.06)
```

```python
def load_data(mock_no,use_FKP=1):
    infile = '/projects/QUIJOTE/Oliver/nseries_mocks/CutskyN%d.rdzw'%mock_no
    outfile = '/projects/QUIJOTE/Oliver/nseries_mocks/CutskyN%d.xyzw'%mock_no
    ra,dec,z,w_fkp,w_sys = np.loadtxt(infile,unpack=True)
    if use_FKP:
        w = w_fkp*w_sys
    pos = transform.SkyToCartesian(ra,dec,z,cosmo)
    outdat = np.vstack([pos.T,w])
    np.savetxt(outfile,outdat.T)
```

```python
for i in range(100):
    try:
        print("Loading mock %d"%i)
        load_data(i)
    except IOError:
        print("No mock %d!"%i)
```

```python
def save_data(mock_no):
    infile = '/projects/QUIJOTE/Oliver/nseries_mocks/CutskyN%d.xyzw'%mock_no
    outfile = '/projects/QUIJOTE/Oliver/nseries_mocks/nseries%d.data'%mock_no
    dat = np.loadtxt(infile)
    np.savetxt(outfile,dat)
    
    # Also convert to .gz file
    os.system('gzip -f %s'%(outfile))
```

```python
for i in range(100):
    try:
        print("Loading mock %d"%i)
        save_data(i)
    except IOError:
        print("No mock %d!"%i)
```
